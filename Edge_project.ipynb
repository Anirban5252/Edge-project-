```python
# here i can demonstrate that the main properties of quantum mechanics in mathematical treatment.
# sci is a quantum state vector which is defined that how to find any particle in a system which have
# also some energy state that is followed a energy evaluated wall. the mathematical threatment of sci and the energy state E
     #quantum state vector sci,ψ(x, y) = Lx^2 * sin(Lx * nx * πx) * Ly^2 * sin(Ly * ny * πy)
     # Energy of a state E(nx, ny) = (2m * ℏ^2 * π^2) * (Lx^2 * nx^2 + Ly^2 * ny^2)
# Here i plot the distribution of the energy state and also show probability of finding place where the particle can find maximum
# and there also highest energy particle


 # This is 3d plot for eigen state or the equivalent state of wave vector of energy state


import numpy as np # for large number array calling
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constants
hbar = 1.0545718e-34  # Reduced Planck's constant (Joule-second)
m = 9.10938356e-31    # Mass of an electron (kg)
Lx = 1e-9             # Width of the potential well in x-direction (1 nm)
Ly = 2e-9             # Width of the potential well in y-direction (2 nm)
Lz = 3e-9             # Width of the potential well in z-direction (3 nm)

# Define the wavefunction for the (n_x, n_y, n_z)-th state
def psi_nx_ny_nz(x, y, z, nx, ny, nz, Lx, Ly, Lz):
    return (np.sqrt(2 / Lx) * np.sin(nx * np.pi * x / Lx)) * \
           (np.sqrt(2 / Ly) * np.sin(ny * np.pi * y / Ly)) * \
           (np.sqrt(2 / Lz) * np.sin(nz * np.pi * z / Lz))

# Define the energy for the (n_x, n_y, n_z)-th state
def energy_nx_ny_nz(nx, ny, nz, Lx, Ly, Lz, m):
    return (hbar**2 * np.pi**2 / (2 * m)) * (nx**2 / Lx**2 + ny**2 / Ly**2 + nz**2 / Lz**2)

# Generate x, y, and z values within the well
x = np.linspace(0, Lx, 30)
y = np.linspace(0, Ly, 30)
z = np.linspace(0, Lz, 30)
X, Y, Z = np.meshgrid(x, y, z)

# Quantum numbers for the state to visualize
nx, ny, nz = 1, 1, 1

# Compute the wavefunction
psi = psi_nx_ny_nz(X, Y, Z, nx, ny, nz, Lx, Ly, Lz)

# Compute the energy
energy = energy_nx_ny_nz(nx, ny, nz, Lx, Ly, Lz, m)

# Create a 3D scatter plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the wavefunction data as a 3D scatter
sc = ax.scatter(X, Y, Z, c=psi**2, cmap='viridis', marker='o', s=5)
fig.colorbar(sc, ax=ax, shrink=0.5, aspect=10, label=r"$|\psi(x, y, z)|^2$")

# Add labels and title
ax.set_title(f"Wavefunction for (n_x, n_y, n_z)=({nx}, {ny}, {nz})\nEnergy = {energy:.2e} J")
ax.set_xlabel("x (m)")
ax.set_ylabel("y (m)")
ax.set_zlabel("z (m)")

plt.show()
```

```python
# This 2d plot for the wave vector where i use numpy a largest array


import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constants
hbar = 1.0545718e-34  # Reduced Planck's constant (Joule-second)
m = 9.10938356e-31    # Mass of an electron (kg)
Lx = 1e-9             # Width of the potential well in x-direction (1 nm)
Ly = 2e-9             # Width of the potential well in y-direction (2 nm)

# Define the wavefunction for the (n_x, n_y)-th state
def psi_nx_ny(x, y, nx, ny, Lx, Ly):
    return (np.sqrt(2 / Lx) * np.sin(nx * np.pi * x / Lx)) * (np.sqrt(2 / Ly) * np.sin(ny * np.pi * y / Ly))

# Define the energy for the (n_x, n_y)-th state
def energy_nx_ny(nx, ny, Lx, Ly, m):
    return (hbar**2 * np.pi**2 / (2 * m)) * (nx**2 / Lx**2 + ny**2 / Ly**2)

# Generate x and y values within the well
x = np.linspace(0, Lx, 100)
y = np.linspace(0, Ly, 100)
X, Y = np.meshgrid(x, y)

# Quantum numbers for the state to visualize
nx, ny = 1, 1

# Compute the wavefunction
psi = psi_nx_ny(X, Y, nx, ny, Lx, Ly)

# Compute the energy
energy = energy_nx_ny(nx, ny, Lx, Ly, m)

# Plot the wavefunction
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, psi, cmap='viridis', edgecolor='none')

# Add labels and title
ax.set_title(f"Wavefunction for (n_x, n_y)=({nx}, {ny})\nEnergy = {energy:.2e} J")
ax.set_xlabel("x (m)")
ax.set_ylabel("y (m)")
ax.set_zlabel(r"Wavefunction $\psi(x, y)$")

# Show color bar
fig.colorbar(surf, shrink=0.5, aspect=10)

plt.show()
```

```python
# This is 1d plot where i write the code in very simple

import numpy as np
import matplotlib.pyplot as plt

# Constants
hbar = 1.0545718e-34  # Reduced Planck's constant (Joule-second)
m = 9.10938356e-31    # Mass of an electron (kg)
L = 1e-9               # Width of the potential well (1 nm)

# Define the wavefunction for the n-th state
def psi_n(x, n, L):
    return np.sqrt(2 / L) * np.sin(n * np.pi * x / L)

# Define the energy for the n-th state
def energy_n(n, L, m):
    return (n**2 * np.pi**2 * hbar**2) / (2 * m * L**2)

# Generate x values within the well
x = np.linspace(0, L, 1000)

# Plot wavefunctions and energy levels
plt.figure(figsize=(10, 6))
for n in range(1, 4):  # Plot the first 3 states
    psi = psi_n(x, n, L)
    plt.plot(x, psi, label=f"n={n}, E={energy_n(n, L, m):.2e} J")

plt.title("Wavefunctions in a 1D Infinite Potential Well")
plt.xlabel("Position x (m)")
plt.ylabel("Wavefunction \u03C8(x)")
plt.legend()
plt.grid()
plt.show()
```
